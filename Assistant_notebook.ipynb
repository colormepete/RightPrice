```python
import pandas as pd
import requests
```

```python
def get_keys_from_file(filename):
    import imp
    f = open(filename)
    global keys
    keys = imp.load_source('keys',filename)
    f.close()
    
get_keys_from_file('/Users/peterschnatz/Insight/Project1/RightPrice/eventbrite_keys.py')

eventbrite_api_token = keys.PRIVATE_TOKEN

eventbrite_api_base = keys.base
```

```python
request = requests.get(eventbrite_api_base+'subcategories/?token='+eventbrite_api_token)
request_json = request.json()
request_json
```

```python
cat_dict={}

for i in range(1,5):
    request = requests.get(eventbrite_api_base+'subcategories/?page='+str(i)+'&token='+eventbrite_api_token)
    request_json = request.json()

    for cat in request_json['subcategories']:
        cat_dict.update({cat['id']:cat['name']})

print(cat_dict)
```
